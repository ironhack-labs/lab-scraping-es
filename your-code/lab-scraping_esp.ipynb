#### Desafío 2 - Mostrar los repositorios de Python en tendencia en GitHub

Los pasos para resolver este problema son similares al anterior, excepto que necesitas encontrar los nombres de los repositorios en lugar de los nombres de los desarrolladores.

````python
url2 = 'https://github.com/trending/python?since=daily'
response2 = requests.get(url2)
soup2 = BeautifulSoup(response2.text, 'html.parser')
````

````python
import re

repositories = soup2.find_all('h2','h3 lh-condensed')

def repo_getter(repositories):
    repo_list = []
    for repo in repositories:
        repo_list.append((re.sub(r'\s+',' ',repo.get_text())).strip())
    return(repo_list)


final_repos = repo_getter(repositories)

final_repos

````

````python
names = []
a = soup2.find_all(class_='h3 lh-condensed')
for item in a:
    result = item.find_all('a')
    elementos = result[0].get_text().strip().split(' ')
    names.append(elementos[-1])
print(names)
````

#### Desafío 4 - Recuperar todos los enlaces a páginas en Wikipedia que se refieren a algún tipo de Python.

````python
url4 ='https://en.wikipedia.org/wiki/Python' 
python = requests.get(f"{url4}").text
soup4 = BeautifulSoup(python, 'html.parser')
````

````python
python_links = soup4.find_all('a')

python_links
    
````

````python
def get_python_links(links):
    final_links = []
    for link in links:
        if 'python' in (str(link.get('href'))).lower():
            final_links.append(link.get('href'))
    return(final_links)

get_python_links(python_links)
            
````

````python
movie_titles = soup11.find_all('h3', class_ = 'ipc-title__text')
movie_date = soup11.find_all('span', class_ = 'sc-b189961a-8 kLaxqf cli-title-metadata-item')
ratings = [span['aria-label'] for span in soup11.find_all('span', class_ ='ipc-rating-star ipc-rating-star--base ipc-rating-star--imdb ratingGroup--imdb-rating')]
directors = soup11.find_all('div', class_ ='sc-74bf520e-3 klvfeN dli-parent')

directors

#ratings
rating_list = [(rating.split(': ')[1]) for rating in ratings]


#movies
movie_list = [movie.get_text().split('. ')[1].strip() for movie in movie_titles if len(movie.get_text().split('. ')) > 1]

#movie dates
movie_dates = [date.get_text().strip() for date in movie_date]
movie_dates = [date for date in movie_dates if date.isdigit() and len(date)==4]


combined_list = [(movie, date, rating) for movie, date, rating in zip(movie_list,movie_dates, rating_list)]


import pandas as pd
top250df = pd.DataFrame(combined_list, columns=['Movie','Date','Rating'])

top250df
````

#### Desafío 14 - Nombre del libro, precio y disponibilidad de stock como un dataframe de pandas.
